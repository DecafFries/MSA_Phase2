## 2. Clean data

#### 2.1 Merge All Datasets

```python
w_ft.Date = pd.to_datetime(w_ft.Date)
w_ft.Date
```

```python
w_sale.Date = pd.to_datetime(w_sale.Date)
w_sale.Date
```

```python
w_df = w_sale.merge(w_ft, how='left', on=['Store', 'Date', 'IsHoliday'])
w_df = w_df.merge(w_st, how='left', on='Store')
w_df.head(10)
```

```python
w_df.tail(10)
```

```python
w_df.info()
```

#### 2.2 Handling Missing Values

```python
# Check for remaining missing values in both dataframes
print("Missing values in W Store DataFrame:")
print(w_df.isnull().sum())
```

```python
# Fill missing values for w_df
w_df.fillna({
    'MarkDown1': 0,
    'MarkDown2': 0,
    'MarkDown3': 0,
    'MarkDown4': 0,
    'MarkDown5': 0,
}, inplace=True)

# Check for remaining missing values in both dataframes
print("Missing values in W Store DataFrame:")
print(w_df.isnull().sum())
```

#### 2.3 Remove Outliers

```python
# Removing outliers for W Store
w_df = w_df[(np.abs(stats.zscore(w_df.select_dtypes(include=[np.number]))) < 3).all(axis=1)]
```

#### 2.4 Encoding Categorical Variables

```python
w_df.IsHoliday.unique()
```

```python
w_df.Type.unique()
```

```python
# Convert boolean column IsHoliday to integers
w_df.IsHoliday= w_df.IsHoliday.astype(int)
```

```python
# One-hot encode the Type column
w_df = pd.get_dummies(w_df, columns=['Type'], drop_first=True)
```

```python
w_df.Type_B= w_df.Type_B.astype(int)
w_df.Type_C= w_df.Type_C.astype(int)
```

```python
w_df.head()
```

```python
# Aggregate W store data
w_sales_weekly = w_df.resample('W-Mon', on='Date').sum().reset_index().sort_values(by='Date')
w_sales_monthly = w_df.resample('ME', on='Date').sum().reset_index().sort_values(by='Date')
w_sales_quarterly = w_df.resample('QE', on='Date').sum().reset_index().sort_values(by='Date')
w_sales_yearly = w_df.resample('YE', on='Date').sum().reset_index().sort_values(by='Date')
```

```python
w_df.head()
```

## 3. Visualise data

```python
# Visualize W Store Data
# Line plot for W store data - Weekly
plt.figure(figsize=(12, 6))
sns.lineplot(data=w_sales_weekly, x='Date', y='Weekly_Sales')
plt.title('Combined Weekly Sales for All Stores and Departments')
plt.xlabel('Date')
plt.ylabel('Weekly Sales')
plt.show()

# Line plot for W store data - Monthly
plt.figure(figsize=(12, 6))
sns.lineplot(data=w_sales_monthly, x='Date', y='Weekly_Sales')
plt.title('Combined Monthly Sales for All Stores and Departments')
plt.xlabel('Date')
plt.ylabel('Monthly Sales')
plt.show()

# Line plot for W store data - Quarterly
plt.figure(figsize=(12, 6))
sns.lineplot(data=w_sales_quarterly, x='Date', y='Weekly_Sales')
plt.title('Combined Quarterly Sales for All Stores and Departments')
plt.xlabel('Date')
plt.ylabel('Quarterly Sales')
plt.show()

# Line plot for W store data - Yearly
plt.figure(figsize=(12, 6))
sns.lineplot(data=w_sales_yearly, x='Date', y='Weekly_Sales')
plt.title('Combined Yearly Sales for All Stores and Departments')
plt.xlabel('Date')
plt.ylabel('Yearly Sales')
plt.show()
```

## 4. Identify correlated variables

```python
# Compute the correlation matrix
corr_matrix = w_df.corr()

# Display the correlation matrix
print(corr_matrix)
```

```python
# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Draw the heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', vmin=-1, vmax=1)

# Add titles and labels
plt.title('Correlation Matrix of Features', fontsize=15)
plt.show()
```

```python
# Find pairs of highly correlated variables
high_corr_pairs = [(var1, var2, corr_matrix.loc[var1, var2]) for var1 in corr_matrix.columns for var2 in corr_matrix.columns if var1 != var2 and abs(corr_matrix.loc[var1, var2]) > 0.75]

# Display highly correlated pairs
print("Highly Correlated Pairs:")
for var1, var2, corr in high_corr_pairs:
    print(f"{var1} and {var2}: {corr}")
```
